SMART QUESTION
Apakah ada pola dalam sentimen tweet berdasarkan waktu tweet atau usia pengguna di negara tertentu?

Data Wrangling
Sudah Dilakukan saat pembersihan data

Eksplorasi data

```python
# Mengeksplorasi distribusi sentimen berdasarkan waktu tweet
sentiment_time = df.groupby(['Time of Tweet', 'sentiment']).size().unstack().fillna(0)

# Mengeksplorasi distribusi sentimen berdasarkan usia pengguna
sentiment_age = df.groupby(['Age of User', 'sentiment']).size().unstack().fillna(0)

# Menampilkan hasil
sentiment_time, sentiment_age
```

Visualisasi Data

```python
# Visualisasi distribusi sentimen berdasarkan waktu tweet
plt.figure(figsize=(12, 6))
sentiment_time.plot(kind='bar', stacked=True)
plt.title('Distribusi Sentimen berdasarkan Waktu Tweet')
plt.xlabel('Waktu Tweet')
plt.ylabel('Jumlah Tweet')
plt.xticks(rotation=45)
plt.legend(title='Sentimen')
plt.show()

# Visualisasi distribusi sentimen berdasarkan usia pengguna
plt.figure(figsize=(12, 6))
sentiment_age.plot(kind='bar', stacked=True)
plt.title('Distribusi Sentimen berdasarkan Usia Pengguna')
plt.xlabel('Usia Pengguna')
plt.ylabel('Jumlah Tweet')
plt.xticks(rotation=45)
plt.legend(title='Sentimen')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Define columns
categorical_cols = ['sentiment', 'Time of Tweet', 'Age of User', 'Country']
numerical_cols = ['Population -2020', 'Land Area (Km²)', 'Density (P/Km²)']

# Create transformers
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
numerical_transformer = StandardScaler()

# Create preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Apply transformations to the dataset
preprocessed_data = preprocessor.fit_transform(df)

# Convert to DataFrame
preprocessed_df = pd.DataFrame(preprocessed_data.toarray())
preprocessed_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df and numerical_cols are defined
# Create histograms for numerical features
numerical_data = df[numerical_cols]
numerical_data.hist(bins=15, figsize=(15, 6), layout=(2, 3))
plt.tight_layout()
plt.show()

# Create bar plot for 'sentiment' feature
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='sentiment')
plt.tight_layout()
plt.show()

# Create correlation matrix heatmap
corr_matrix = numerical_data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
