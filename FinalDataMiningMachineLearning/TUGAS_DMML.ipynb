```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
import joblib

# Muat data
data = pd.read_csv('test_bersih.csv')
texts = data['text']  # Kolom teks
sentiments = data['sentiment']  # Kolom sentimen

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(texts, sentiments, test_size=0.2, random_state=42)

# Membuat pipeline untuk vektorisasi dan klasifikasi dengan Logistic Regression
logreg_pipeline = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('classifier', LogisticRegression())
])

# Melatih model Logistic Regression
logreg_pipeline.fit(X_train, y_train)

# Menyimpan model Logistic Regression
joblib.dump(logreg_pipeline, 'logreg_sentiment_model.pkl')

# Membuat pipeline untuk vektorisasi dan klasifikasi dengan SVM
svm_pipeline = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('classifier', SVC())
])

# Melatih model SVM
svm_pipeline.fit(X_train, y_train)

# Menyimpan model SVM
joblib.dump(svm_pipeline, 'svm_sentiment_model.pkl')

# Membuat pipeline untuk vektorisasi dan klasifikasi dengan Naive Bayes
nb_pipeline = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('classifier', MultinomialNB())
])

# Melatih model Naive Bayes
nb_pipeline.fit(X_train, y_train)

# Menyimpan model Naive Bayes
joblib.dump(nb_pipeline, 'naive_bayes_sentiment_model.pkl')

# Membuat pipeline untuk vektorisasi dan klasifikasi dengan MLP
mlp_pipeline = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('classifier', MLPClassifier(max_iter=1000))
])

# Melatih model MLP
mlp_pipeline.fit(X_train, y_train)

# Menyimpan model MLP
joblib.dump(mlp_pipeline, 'mlp_sentiment_model.pkl')
```

```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Muat data
data = pd.read_csv('test_bersih.csv')
texts = data['text']  # Kolom teks
sentiments = data['sentiment']  # Kolom sentimen

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(texts, sentiments, test_size=0.2, random_state=42)

# Load models
logreg_pipeline = joblib.load('logreg_sentiment_model.pkl')
svm_pipeline = joblib.load('svm_sentiment_model.pkl')
nb_pipeline = joblib.load('naive_bayes_sentiment_model.pkl')
mlp_pipeline = joblib.load('mlp_sentiment_model.pkl')

# Make predictions
logreg_preds = logreg_pipeline.predict(X_test)
svm_preds = svm_pipeline.predict(X_test)
nb_preds = nb_pipeline.predict(X_test)
mlp_preds = mlp_pipeline.predict(X_test)

# Calculate classification reports
logreg_report = classification_report(y_test, logreg_preds, output_dict=True)
svm_report = classification_report(y_test, svm_preds, output_dict=True)
nb_report = classification_report(y_test, nb_preds, output_dict=True)
mlp_report = classification_report(y_test, mlp_preds, output_dict=True)

# Calculate confusion matrices
logreg_cm = confusion_matrix(y_test, logreg_preds)
svm_cm = confusion_matrix(y_test, svm_preds)
nb_cm = confusion_matrix(y_test, nb_preds)
mlp_cm = confusion_matrix(y_test, mlp_preds)

# Plotting function for confusion matrix
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(title)
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()

# Plot confusion matrices
plot_confusion_matrix(logreg_cm, 'Logistic Regression Confusion Matrix')
plot_confusion_matrix(svm_cm, 'SVM Confusion Matrix')
plot_confusion_matrix(nb_cm, 'Naive Bayes Confusion Matrix')
plot_confusion_matrix(mlp_cm, 'MLP Confusion Matrix')

# Display classification reports
print("Logistic Regression Classification Report:")
print(pd.DataFrame(logreg_report).transpose())

print("SVM Classification Report:")
print(pd.DataFrame(svm_report).transpose())

print("Naive Bayes Classification Report:")
print(pd.DataFrame(nb_report).transpose())

print("MLP Classification Report:")
print(pd.DataFrame(mlp_report).transpose())
```

```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Muat data
data = pd.read_csv('test_bersih.csv')
texts = data['text']  # Kolom teks
sentiments = data['sentiment']  # Kolom sentimen

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(texts, sentiments, test_size=0.2, random_state=42)

# Load models
logreg_pipeline = joblib.load('logreg_sentiment_model.pkl')
svm_pipeline = joblib.load('svm_sentiment_model.pkl')
nb_pipeline = joblib.load('naive_bayes_sentiment_model.pkl')
mlp_pipeline = joblib.load('mlp_sentiment_model.pkl')

# Make predictions
logreg_preds = logreg_pipeline.predict(X_test)
svm_preds = svm_pipeline.predict(X_test)
nb_preds = nb_pipeline.predict(X_test)
mlp_preds = mlp_pipeline.predict(X_test)

# Calculate classification reports
logreg_report = classification_report(y_test, logreg_preds, output_dict=True)
svm_report = classification_report(y_test, svm_preds, output_dict=True)
nb_report = classification_report(y_test, nb_preds, output_dict=True)
mlp_report = classification_report(y_test, mlp_preds, output_dict=True)

# Extract metrics for each model
models = ['Logistic Regression', 'SVM', 'Naive Bayes', 'MLP']
accuracy = [
    logreg_report['accuracy'],
    svm_report['accuracy'],
    nb_report['accuracy'],
    mlp_report['accuracy']
]
precision = [
    logreg_report['weighted avg']['precision'],
    svm_report['weighted avg']['precision'],
    nb_report['weighted avg']['precision'],
    mlp_report['weighted avg']['precision']
]
recall = [
    logreg_report['weighted avg']['recall'],
    svm_report['weighted avg']['recall'],
    nb_report['weighted avg']['recall'],
    mlp_report['weighted avg']['recall']
]
f1_score = [
    logreg_report['weighted avg']['f1-score'],
    svm_report['weighted avg']['f1-score'],
    nb_report['weighted avg']['f1-score'],
    mlp_report['weighted avg']['f1-score']
]

# Create a DataFrame for plotting
metrics_df = pd.DataFrame({
    'Model': models,
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1 Score': f1_score
})

# Plotting the metrics
metrics_df.set_index('Model', inplace=True)
metrics_df.plot(kind='bar', figsize=(10, 6))
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
